```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
gapminder = pd.read_csv('gapminderData.csv')
```

```python
gapminder.head()
```

```python
data2007 = gapminder[gapminder.year == 2007]
data2007.head()
```

```python
plt.scatter(data2007.gdpPercap, data2007.lifeExp)
plt.xscale('log')
plt.show()
```

```python
gapminder.year.unique()
```

```python
gapminder.dtypes
    
```

```python
gapminder.continent = pd.Categorical(gapminder.continent)
gapminder.dtypes
```

```python
for i in gapminder.year.unique():
    tmp = gapminder[gapminder.year == i]
    plt.hist(tmp['lifeExp'], bins=20)
    plt.show()
    plt.xlabel('Life Expectancy')
    plt.title('Year: ' + str(i))
```

```python
for i in gapminder.year.unique():
    tmp = gapminder[gapminder.year == i]
    plt.scatter(tmp['gdpPercap'], tmp['lifeExp'])
    plt.xscale('log')
    plt.xlabel('GDP Per Capita [USD]')
    plt.ylabel('Life Expectancy [Years]')
    plt.title('World Development in ' + str(i))
    plt.show()
```

```python
for i in gapminder.year.unique():
    tmp = gapminder[gapminder.year == i]
    plt.scatter(tmp['gdpPercap'], tmp['lifeExp'])
    plt.xscale('log')
    plt.xlabel('GDP Per Capita [USD]')
    plt.ylabel('Life Expectancy [Years]')
    plt.title('World Development in ' + str(i))
    plt.xticks([1000, 10000, 100000], ['1k', '10k', '100k'])
    plt.show()
```

```python
for i in gapminder.year.unique():
    tmp = gapminder[gapminder.year == i]
    sns.scatterplot(tmp['gdpPercap'],
                   tmp['lifeExp'],
                   hue=gapminder['continent'],)
    plt.xscale('log')
    plt.xlabel('GDP Per Capita [USD]')
    plt.ylabel('Life Expectancy [Years]')
    plt.title('World Development in ' + str(i))
    plt.xticks([1000, 10000, 100000], ['1k', '10k', '100k'])
    plt.show()
```

```python
for i in gapminder.year.unique():
    tmp = gapminder[gapminder.year == i]
    sns.scatterplot(tmp['gdpPercap'],
                   tmp['lifeExp'],
                   hue=gapminder['continent'],
                   s=tmp['pop']/1000000,
                   alpha=0.5)
    plt.xscale('log')
    plt.xlabel('GDP Per Capita [USD]')
    plt.ylabel('Life Expectancy [Years]')
    plt.title('World Development in ' + str(i))
    plt.xticks([1000, 10000, 100000], ['1k', '10k', '100k'])
    plt.show()
```
